```python
price_df = pd.read_csv('autoscout24-germany-dataset.csv')
```

```python
missing_values = data.isna().sum()
print(missing_values)
```

```python
import pandas as pd
import re

def normalize_name(s):
    """Нормализация названий с сохранением структуры для частичных совпадений"""
    s = str(s).lower().strip()
    s = re.sub(r'[-_]', ' ', s)  # Заменяем дефисы и подчеркивания на пробелы
    s = re.sub(r'\s+', ' ', s)   # Удаляем множественные пробелы
    return s


# Нормализация названий
data['mk_norm'] = data['Mk'].apply(normalize_name)
data['cn_norm'] = data['Cn'].apply(normalize_name)

price_df['make_norm'] = price_df['make'].apply(normalize_name)
price_df['model_norm'] = price_df['model'].apply(normalize_name)

# Создаем словарь для поиска цен
price_dict = price_df.groupby(['make_norm', 'model_norm'])['price'].mean().to_dict()

def find_best_match(target_make, target_model):
    """Поиск лучшего совпадения модели"""
    matches = []
    
    for (make, model), price in price_dict.items():
        if make == target_make:
            # Проверяем частичное совпадение
            if target_model in model or model in target_model:
                matches.append(price)
    
    return int(sum(matches)/len(matches)) if matches else None

# Добавляем цены с учетом частичных совпадений
data['price'] = data.apply(
    lambda row: find_best_match(row['mk_norm'], row['cn_norm']), 
    axis=1
)

# Фильтрация и удаление дубликатов
result = (
    data
    .dropna(subset=['price'])  # Удаляем строки без цены
    .drop_duplicates(subset=['Mk', 'Cn'], keep='first')  # Уникальные комбинации
     # Выбираем нужные колонки
)

```

```python
np.random.seed(42)  # Для воспроизводимости

# Создаем значения с нормальным распределением (среднее = 2.6, std = 0.3)
frontal_area = np.random.normal(loc=2.6, scale=0.3, size=len(result))

# Обрезаем значения до диапазона [2.0, 3.2] и округляем
frontal_area = np.clip(frontal_area, 2.0, 3.2).round(1)

# Добавляем столбец в DataFrame
result["frontal_area_m2"] = frontal_area
result.to_csv('data_with_prices.csv', index=False)
print("Данные успешно сохранены в data_with_prices.csv")
```
